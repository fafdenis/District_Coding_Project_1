```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
total_migrants_csv="UN_MigrantStockByAge_2017(total data).csv"
female_migrants_csv="UN_MigrantStockByAge_2017(female data).csv"
male_migrants_csv="UN_MigrantStockByAge_2017(male data).csv"
```

```python
total_migrants_df=pd.read_csv(total_migrants_csv)
female_migrants_df=pd.read_csv(female_migrants_csv)
male_migrants_df=pd.read_csv(male_migrants_csv)
#total_migrants_df
```

```python
#total_migrants_df
us_tmigrants= total_migrants_df.loc[total_migrants_df["Major area, region, country or area of destination"] == "United States of America", :]
```

```python
us_tmigrants
```

```python
us_F_migrants= female_migrants_df.loc[female_migrants_df["Major area, region, country or area of destination"] == "United States of America", :]
us_F_migrants.drop(['14-Oct','9-May'],axis=1)
us_F_migrants
```

```python
us_M_migrants= male_migrants_df.loc[male_migrants_df["Major area, region, country or area of destination"] == "United States of America", :]
us_M_migrants
```

```python
#us_tmigrants.drop(columns= ["Gender"], axis=1)
```

```python
us_tmigrants.insert(0, "Gender", 'Both', allow_duplicates=False)
#us_tmigrants
```

```python
us_tmigrants
```

```python
us_F_migrants.insert(0, "Gender", 'Female', allow_duplicates=False)
us_F_migrants.head()
```

```python
us_M_migrants.insert(0, "Gender", 'Male', allow_duplicates=False)
us_M_migrants.head()
```

```python
US_migrant_df=pd.concat( [us_tmigrants, us_F_migrants, us_M_migrants])
US_migrant_df.head()
```

```python
#bin= [0, 9, 19, 29, 39, 49, 59, 69, 79]
#group_labels=["<10", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70+"]
#age_df=pd.cut(US_migrant_df["Age"], bins, labels=group_labels)
#age_df
```

```python
bins= [1989,1999,2009,2017]
decade_names=["Late 90's", "Early 2000's", "Current Decade"]
US_migrant_df['Decades']=pd.cut(US_migrant_df["Year"], bins, labels=decade_names)

US_migrant_df.head()
```

```python
numeric_cols = [col for col in US_migrant_df.columns if "-" in col]
for col in numeric_cols:
    US_migrant_df[col]= US_migrant_df[col].str.replace(",","").astype(float)
    
US_migrant_df.dtypes
```

```python
US_migrant_df['75+']= US_migrant_df['75+'].str.replace(",","").astype(float)
```

```python
US_migrant_df['Total']= US_migrant_df['Total'].str.replace(",","").astype(float)
```

```python
decade_group=US_migrant_df.groupby('Decades')["Total"].sum()
decade_group.plot(kind='bar', color= 'orange', width= 0.3)
plt.ylabel('Number of Migrants by the Millions')
plt.title('US Migration Over Time')
plt.tight_layout()
plt.savefig("group_project decade3")
```
